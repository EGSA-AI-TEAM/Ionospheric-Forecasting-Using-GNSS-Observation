```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as mcolors

df = pd.read_csv('2021-2024.csv')
print(df['datetime'].dtype)
df['datetime'] = pd.to_datetime(df['datetime'])
print(df['datetime'].dtype)
```

```python
def bar_color(value):
    if 0 <= value <= 3:
        return 'green'
    elif 3 < value < 5:
        return 'yellow'
    elif 5 <= value <= 10:
        return 'red'
    else:
        return 'black'

fig, (ax1, ax2,ax4, ax5, ax6, ax7) = plt.subplots(6, 1, figsize=(40, 30))
fig.suptitle('Geomagneic Data 20240101-20240508',fontsize=30, fontweight='bold',fontfamily='serif')
ax1.plot(df['datetime'],df['BZ, nT (GSM)'], label='BZ, nT (GSM)')
ax1.set_title('BZ',loc = 'center',fontsize=20,fontfamily='serif')
ax1.set_ylabel('nT (GSM)')
ax2.plot(df['datetime'],df['SW Proton Density, N/cm^3'], label='SW Proton Density, N/cm^3')
ax2.set_title('SW Proton Density',fontsize=20,fontfamily='serif')
ax2.set_ylabel('N/cm^3')
# ax3.plot(df['datetime'],df['SW Plasma Speed, km/s'], label='SW Plasma Speed, km/s')
# ax3.set_title('SW Plasma Speed, km/s')
colors = df['Kp index'].apply(bar_color)
ax4.bar(df['datetime'],df['Kp index'], color= colors, label='Kp index',edgecolor='black', width=0.05, linewidth=0.03)
ax4.set_title('Kp index',fontsize=20,fontfamily='serif')
ax5.plot(df['datetime'],df['Dst-index, nT'], label='Dst-index, nT')
ax5.set_title('Dst-index',fontsize=20,fontfamily='serif')
ax5.set_ylabel('nT')
ax6.plot(df['datetime'],df['f10.7_index'], label='f10.7_index')
ax6.set_title('f10.7_index',fontsize=20,fontfamily='serif')
ax7.plot(df['datetime'],df['R (Sunspot No.)'], label='R (Sunspot No.)')
ax7.set_title('R (Sunspot No.)',fontsize=20,fontfamily='serif')

plt.tight_layout(rect=[0, 0, 1, 0.98])
ax1.legend(fontsize = 20)
ax1.grid(color='grey', linestyle='dashdot', linewidth=0.5)
ax2.legend(fontsize = 20)
ax2.grid(color='grey', linestyle='dashdot', linewidth=0.5)
# ax3.legend()
# ax3.grid()
ax4.legend(fontsize = 20)
ax4.grid(color='grey', linestyle='dashdot', linewidth=0.5)
ax5.legend(fontsize = 20)
ax5.grid(color='grey', linestyle='dashdot', linewidth=0.5)
ax6.legend(fontsize = 20)
ax6.grid(color='grey', linestyle='dashdot', linewidth=0.5)
ax7.legend(fontsize = 20)
ax7.grid(color='grey', linestyle='dashdot', linewidth=0.5)

```
