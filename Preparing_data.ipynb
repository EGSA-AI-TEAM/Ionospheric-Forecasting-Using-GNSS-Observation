```python
import os
import pandas as pd
from tqdm import tqdm

# Define converters for each column except the first two
converters = {
    "PRN": lambda x: pd.to_numeric(x, errors='coerce'),
    "Az": lambda x: pd.to_numeric(x, errors='coerce'),
    "Ele": lambda x: pd.to_numeric(x, errors='coerce'),
    "Lat": lambda x: pd.to_numeric(x, errors='coerce'),
    "Lon": lambda x: pd.to_numeric(x, errors='coerce'),
    "Stec": lambda x: pd.to_numeric(x, errors='coerce'),
    "Vtec": lambda x: pd.to_numeric(x, errors='coerce'),
    "S4": lambda x: pd.to_numeric(x, errors='coerce')
}

# Specify the path to your first file
file_path = f'AI CARO\\CARO001-{year}-01-01.Cmn'

# Read the initial file with updated parameters
df = pd.read_csv(file_path, sep='\s+', header=None, names=["MJdate", "Time", "PRN", "Az", "Ele", "Lat", "Lon", "Stec", "Vtec", "S4"],
                 dtype={"MJdate": str, "Time": str}, converters=converters, low_memory=False)

# List to keep track of processed files
done = [file_path]

# Loop through all files in the directory
for i in tqdm(os.listdir('AI CARO')):
    if year in i and i not in done:
        d = pd.read_csv(os.path.join('AI CARO', i), sep='\s+', header=None, names=["MJdate", "Time", "PRN", "Az", "Ele", "Lat", "Lon", "Stec", "Vtec", "S4"],
                        dtype={"MJdate": str, "Time": str}, converters=converters, low_memory=False)
        df = pd.concat([df, d], axis=0, ignore_index=True)
        done.append(i)
```

```python
import pandas as pd
import datetime


# Function to convert Julian date to Gregorian date
# Function to convert Julian date to Gregorian date
def julian_to_gregorian(julian_date):
    return datetime.datetime(1858, 11, 17) + datetime.timedelta(days=julian_date)

# Ensure MJdate is numeric and apply the function to convert
df['MJdate'] = pd.to_numeric(df['MJdate'], errors='coerce')
df['GregorianDate'] = df['MJdate'].apply(julian_to_gregorian)

# Check the type of 'GregorianDate' before conversion
print("Before conversion:")
print(df['GregorianDate'].dtype)

# Convert GregorianDate to datetime type
df['GregorianDate'] = pd.to_datetime(df['GregorianDate'])

# Check the type of 'GregorianDate' after conversion
print("After conversion:")
print(df['GregorianDate'].dtype)

# Format the GregorianDate to show only time up to seconds
df['GregorianDate'] = df['GregorianDate'].dt.strftime('%Y-%m-%d %H:%M:%S')

# Display the DataFrame
df
```
